```python
# create a gif that displays the euler method on f'(x)=f(x) with f(0)=1
import matplotlib.pyplot as plt
import numpy as np

custom_color = (0/255, 137/255, 68/255)  # Custom color for the plot


STEP_SIZE = 1/100
x = np.linspace(0, 2, 1000)
y_true = np.exp(x)
plt.plot(x, y_true, 'lightcoral')

x_euler = np.linspace(0, 2, int(1/STEP_SIZE))
y_euler = []
y_euler.append(1)
for i in range(1,int(1/STEP_SIZE)):
    y_euler.append(y_euler[-1] + y_euler[-1]*STEP_SIZE*2)



plt.plot(x_euler, y_euler, 'darkgreen')

# change face color 
# plt.gca().set_facecolor('black')
# # set grid color
plt.grid(color = custom_color, linestyle = '--', linewidth = 0.5, alpha = 0.3)
ax = plt.gca()

ax.spines['bottom'].set_color(custom_color)
ax.spines['top'].set_color(custom_color)
ax.spines['right'].set_color(custom_color)
ax.spines['left'].set_color(custom_color)
ax.yaxis.label.set_color(custom_color)
ax.xaxis.label.set_color(custom_color)
# ticks
ax.tick_params(axis='x', colors=custom_color)
ax.tick_params(axis='y', colors=custom_color)

plt.title('Euler method on f\'(x)=f(x) with f(0)=1', color=custom_color, fontweight='bold')
plt.legend(['True solution', 'Euler method'])


plt.show()
len(x_euler)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
num_simulations = 10
num_steps = 1000

# Initialize the Brownian motion
brownian_motion = np.zeros((num_simulations, num_steps))


# Generate the Brownian motion
for i in range(num_simulations):
    for j in range(num_steps):
        brownian_motion[i, j] = brownian_motion[i, j-1] + np.random.normal(0, np.sqrt(1/num_steps))


# Plot the Brownian motion
plt.figure(figsize=(10, 6))
for i in range(num_simulations):
    if i < num_simulations/3:
        color = plt.cm.Greens(0.5 + 0.5 * i/(num_simulations/3))
    elif i < 2*num_simulations/3:
        color = plt.cm.GnBu(0.4 + 0.6 * (i-num_simulations/3)/(num_simulations/3))
    else:
        color = plt.cm.YlGn(0.4 + 0.6 * (i-2*num_simulations/3)/(num_simulations/3))
    
    alpha = 0.6 + (0.4 * i / (num_simulations - 1))  # Keep lines more visible
    plt.plot(brownian_motion[i], color=color, alpha=alpha)

plt.grid(color = custom_color, linestyle = '--', linewidth = 0.5, alpha = 0.3)
ax = plt.gca()

ax.spines['bottom'].set_color(custom_color)
ax.spines['top'].set_color(custom_color)
ax.spines['right'].set_color(custom_color)
ax.spines['left'].set_color(custom_color)
ax.yaxis.label.set_color(custom_color)
ax.xaxis.label.set_color(custom_color)
# ticks
ax.tick_params(axis='x', colors=custom_color)
ax.tick_params(axis='y', colors=custom_color)

plt.title('Brownian Motion Simulations', color=custom_color, fontweight='bold')
plt.xlabel('Time')
plt.ylabel('Position')
```
