```python
import numpy as np
import matplotlib.pyplot as plt

def random_density(x, seed=20):
    """Density-like function using a polynomial with randomized coefficients."""
    np.random.seed(seed)
    poly = np.sin(x)+np.sin(0.5*x**2)+np.sin(1/3 * x**3)
    density = np.exp(poly)
    return density / 1.4346 # Ensure non-negative values

# Define the range and calculate the function
x = np.linspace(3, 5, 1000)
y = random_density(x)

custom_color = (0/255, 137/255, 68/255)  # Custom color for the plot

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Randomized Polynomial Density", color=custom_color)
plt.title("$\\bf{p_{data}}(x)$", color=custom_color)

ax = plt.gca()
ax.spines['top'].set_color(custom_color)
ax.spines['bottom'].set_color(custom_color)
ax.spines['left'].set_color(custom_color)
ax.spines['right'].set_color(custom_color)
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
plt.show()
```

```python
# add gaussian noise to the function

def normal_density(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

def q_sigma(sigma):
    all_x = np.linspace(3, 5, 1000)
    all_y = random_density(all_x)
    all_q = np.zeros_like(all_y)
    for i, x in enumerate(all_x):
        all_q[i] = np.sum(normal_density(all_x, x, sigma) * all_y)
    return all_q

x = np.linspace(3, 5, 1000)

plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
for i, sigma in enumerate([0.01, 0.05, 0.1, 0.15, 0.2, 0.3]):
    plt.subplot(2, 3, i+1)
    plt.plot(x, q_sigma(sigma), label="$q_{"+str(sigma)+"}(x)$", color=custom_color)
    plt.title("$\\bf{p_{"+str(sigma)+"}(x)}$", color=custom_color)
    ax = plt.gca()
    ax.spines['top'].set_color(custom_color)
    ax.spines['bottom'].set_color(custom_color)
    ax.spines['left'].set_color(custom_color)
    ax.spines['right'].set_color(custom_color)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    
plt.show()
# for sigma in [0.01, 0.05, 0.1, 0.2, 0.35]:
#     all_q = np.zeros_like(all_y)
#     for i, x in enumerate(all_x):
#         all_q[i] = np.sum(normal_density(x, all_x, sigma) * all_y)

#     plt.figure(figsize=(10, 6))
#     plt.plot(all_x, all_q, label="Randomized Polynomial Density")
#     plt.title(f"Density Function with sigma={sigma}")
#     # plt.gca().axes.get_xaxis().set_visible(False)




```
